### Location Visualization

```python
#LONGITUDE VS LATTITUDE SCATTER PLOTE 
#we only consider longitude value [-180,180] and latitude value[-90,90]

longitude = pd.DataFrame(aviationRawData[abs(aviationRawData['longitude'])<180])
latitude = pd.DataFrame(aviationRawData[abs(aviationRawData['longitude'])<90])

maps = pd.concat([latitude,longitude], axis = 1)
longitude.describe()

f, axes = plt.subplots(1, 1, figsize=(16,8))
plt.scatter(data = maps, x = 'longitude', y = 'latitude')
plt.title("World Map Longitude vs Latitude Scatter Plot")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
```

```python
#SCATTER PLOT FOR LOCATION AROUND UNITED STATES

us = pd.DataFrame(aviationRawData[aviationRawData['country'] == "United States"])
us = pd.DataFrame(us[abs(us['longitude'])<180])
us = pd.DataFrame(us[us['longitude']<-70]) 
us = pd.DataFrame(us[us['latitude'] > 25])

f, axes = plt.subplots(1, 1, figsize=(16,8))
plt.scatter(data = us, x = 'longitude', y = 'latitude')
plt.title("United States Longitude vs Latitude Scatter Plot")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
```
